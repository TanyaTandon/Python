```python
%matplotlib inline
import random 
import matplotlib.pyplot as plt
import timeit

l= [100, 1000, 10000, 100000]
random_choice = [1,2] #Choice 1 when word, choice 2 when number 
def randomlist(n):
    random_list = []
    for j in range(n):
        if (random.choice(random_choice) == 1): #randomizing choice of words and numbers
            word = ''        
            for k in range(random.randint(1,100)): #max length of a word being 100 
                word = word + random.choice('abcdefghijklmnopqrstuvwxyz')
            random_list.append(word)
        else:
            random_list.append(random.randint(1,1000))#max size of a number being 1000 
    return (random_list) 

# PROCEDURAL 
def procedural(random_list,a): 
    count_num = 0 
    for i in random_list:
        if type(i) == int:
            count_num += 1
    #  Number of character elements = Total - number of number elements 
    print("Procedural:The number of elements as numbers: " + str(count_num))
    print("Procedural:The number of character elements are: " + str(a - count_num)) 

#Functional 
def functional(random_list, b):
    count = 0 
    count = sum([1 for i in random_list if type(i) == int])
    print("Functional:The number of elements as numbers: " + str(count))
    print("Functional:The number of character elements are: " + str(b - count)) 
    
#comparing time of Procedural vs Functional 
c = []
for i in l:
    x = 0
    y = 0
    list = randomlist(i) 
    print("The size of set is "+ str(i))
    x = timeit.timeit(lambda: procedural(list, i), number = 1)
    print("  Time for Procedural of size "+ str(i) + " is: " + str(x))
    
    y = timeit.timeit(lambda: functional(list,i), number = 1)
    print("  Time for Functional of size "+ str(i) + " is: " + str(y))
    c.append(x-y)
    print("       Difference in Procedural - Functional time: " + str(x- y))
    print( "\n \n ")
    
t = ["100","1000","10000", "100000"]

plt.bar(t, c)
plt.xlabel("list size")
plt.ylabel("Procedural - Functional time")
plt.title("Time Comparison")
plt.show()
```
